**Importing IRIS dataset**

```python
data = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
print(data.head())
data = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",header=None)
print(data.head())
data.columns = ["Sepal.Length", "Sepal.Width",  "Petal.Length", "Petal.Width",  "Species"]
print(data.head())
```

## Relationships (bi variate and multi variate analysis)

```python
data.plot.scatter(x='Sepal.Length', y='Petal.Length')
```

```python
pd.scatter_matrix(data[["Sepal.Length", "Sepal.Width",  "Petal.Length"]])
plt.show()
pd.scatter_matrix(data[["Sepal.Length", "Sepal.Width",  "Petal.Length"]], figsize=(8,8))
plt.show()
```

```python
pd.tools.plotting.scatter_matrix(data, alpha=0.2, figsize=(10, 10), diagonal='hist');
```
